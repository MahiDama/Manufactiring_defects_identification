```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
# load the npz file
data_path = '/content/drive/MyDrive/QI/train_compiled.npz'
train_data   = np.load(data_path, mmap_mode='r')
for k in train_data.files:
  print(k)
```

```python
x_data = train_data['x_data']
y_data = train_data['y_data']
```

```python
pca = PCA(2)  # project from 30 to 2 dimensions
projected = pca.fit_transform(x_data)
print(x_data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=y_data,
            cmap='rainbow')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.show()
plt.savefig('/content/drive/MyDrive/QI/pca_compiled.png')
```
